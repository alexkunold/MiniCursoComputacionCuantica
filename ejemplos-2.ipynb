```python
import numpy as np
from qiskit import QuantumCircuit
from qiskit_aer import AerSimulator
```

```python
qc = QuantumCircuit(5)
```

```python
qc.draw(output='mpl', filename='./qbit5.pdf')
```

```python
qc = QuantumCircuit(6)
qc.x(0)
qc.y(1)
qc.z(2)
qc.h(3)
qc.t(4)
qc.s(5)
```

```python
qc.draw(output='mpl', filename='./gates1qbit.pdf')
```

```python
qc = QuantumCircuit(6)
qc.cx(0,1)
qc.cx(1,2)
qc.cx(2,4)
qc.cx(0,4)
qc.cx(5,1)
```

```python
qc.draw(output='mpl', filename='./gates2qbit.pdf')
```

```python
qc = QuantumCircuit(2)
qc.cx(0,1)
qc.draw(output='mpl', filename='./cnot.pdf')
```

```python
qc = QuantumCircuit(1,1)
qc.measure(0,0)
qc.draw(output='mpl', filename='./meas.pdf')
```

```python
qc= QuantumCircuit(4)
qc.x(0)
qc.y(1)
qc.h(2)
qc.z(3)
qc.cx(0,1)
qc.cx(2,3)
qc.x(0)
qc.cx(1,2)
qc.x(3)
qc.y(2)
qc.h(1)
qc.z(0)

qc.draw(output='mpl', filename='./universality.pdf')
```

```python
ga = qc.to_gate(label='$U$')
```

```python
qcb = QuantumCircuit(4)
qcb.append(ga,[0,1,2,3])
```

```python
qcb.draw(output='mpl', filename='./un.pdf')
```
